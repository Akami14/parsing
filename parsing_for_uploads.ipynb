```python
url = 'https://api.github.com'
user='Akami14'
git = requests.get(f'{url}/users/{user}/repos')
git.ok
```

Запрос выглядит - https://oauth.vk.com/authorize + доп параметры, которые подаются через & такие как 
версия, редирект юрл и клиент айди - это основны параметры для получения токена
 https://oauth.vk.com/authorize?client_id=1&display=page
 &client_id client_id=CLIENT - из нашего приложения цифры
 &redirect_uri=  https://oauth.vk.com/blank.html как для ост стандарло приложений
 &scope=  - отвечает за права токена т.е инфу которую мы будем собирать
 &response_type=token  
 &v=5.131  береем последнию по ссылку - https://vk.com/dev/versions 
 &state=123456   
 По итогу форма выглядит -

&scope=friends
&response_type=token
&v=5.131
&state=123456
